# Excercise 1

```python
def compute_pay(hours, rate):

    regular_hours = 40
    overtime_rate = rate * 1.5

    regular_pay = regular_hours * rate

    if hours > regular_hours:
        overtime_hours = hours - regular_hours
        overtime_pay = overtime_hours * overtime_rate
        pay = regular_pay + overtime_pay
    else:
        pay = regular_pay

    return pay

hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

pay = compute_pay(hours, rate)

print(f"Pay: {pay:.2f}")
```

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

```python
def computegrade(grade):
    if grade >= 0 and grade <= 100:
        if grade >= 0 and grade <= 74:
            print('Your grade is 3.0')
        elif grade >= 75 and grade <= 77:
            print('Your grade is 3.3')
        elif grade >= 78 and grade <= 81:
            print('Your grade is 3.7')
        elif grade >= 82 and grade <= 85:
            print('Your grade is 4.0')
        elif grade >= 86 and grade <= 89:
            print('Your grade is 4.3')
        elif grade >= 90 and grade <= 93:
            print('Your grade is 4.7')
        elif grade >= 94 and grade <= 97:
            print('Your grade is 5.0')
        elif grade >= 98 and grade <= 100:
            print('Your grade is 5.3')
    else:
        print('Invalid input grade')

grade = float(input('Enter Grade: '))

computegrade(grade)
```
