```python
hours = float(input('Enter your Hours : '))
rate = float(input('Enter your Hours : '))
pay = hours*rate
if hours > 40:
  ot = hours -40
  nrate = ot*(rate*1.5)
  npay = 40*rate + nrate
  print('Pay: ',npay)
else :
  print('Pay: ',pay)
```
