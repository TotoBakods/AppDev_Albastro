```python
with open("mboxt-short.txt", "r") as file:
    time_stamps = []

    for line in file:
        if line.startswith("From "):
            words = line.split()
            time_stamp = words[5]
            time_stamps.append(time_stamp)

    for time_stamp in time_stamps:
        print(time_stamp)
```
