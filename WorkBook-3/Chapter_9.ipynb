```python
def count_words(line):
    word_list = line.split()
    word_count = {}
    for word in word_list:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    return word_count

def main():
    line = input("Enter a line of text: ")
    word_count = count_words(line)
    print("\nWord counts:")
    for word, count in word_count.items():
        print(f"{word}: {count}")

if __name__ == "__main__":
    main()
```
