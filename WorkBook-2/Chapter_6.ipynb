Challenge

```python
def process_name(full_name):
    # Strip trailing spaces and capitalize the first letter
    formatted_name = full_name.strip().capitalize()
    # Convert the rest of the name to lowercase
    formatted_name = formatted_name[:1] + formatted_name[1:].lower()
    return formatted_name

def main():
    while True:
        full_name = input("Enter a full name (or 'quit' to exit): ")
        if full_name.lower() == 'quit':
            break
        formatted_name = process_name(full_name)
        print("Formatted name:", formatted_name)

if __name__ == "__main__":
    main()
```
